```python
import matplotlib.pyplot as plt

import numpy as np
import pandas as pd
import scipy.io as sio

from PIL import Image
```

```python
# support functions ---------------------------------------
def plot_n_image(X, n):
    """ plot first n images
    n has to be a square number
    """
    pic_size = int(np.sqrt(X.shape[1]))
    grid_size = int(np.sqrt(n))

    first_n_images = X[:n, :]

    fig, ax_array = plt.subplots(nrows=grid_size, ncols=grid_size,
                                    sharey=True, sharex=True,figsize=(8, 8) )

    for r in range(grid_size):
        for c in range(grid_size):
            ax_array[r, c].imshow(first_n_images[grid_size * r + c].reshape((pic_size, pic_size)),cmap='gray')

            plt.xticks(np.array([]))
            plt.yticks(np.array([]))


# PCA functions ---------------------------------------
def covariance_matrix(X):
    """
    Args:
        X (ndarray) (m, n)
    Return:
        cov_mat (ndarray) (n, n):
            covariance matrix of X
    """
    m = X.shape[0]

    return (X.T @ X) / m


def normalize(X):
    """
        for each column, X-mean / std
    """
    X_copy = X.copy()
    m, n = X_copy.shape

    for col in range(n):
        X_copy[:, col] = (X_copy[:, col] - X_copy[:, col].mean()) / X_copy[:, col].std()

    return X_copy


def pca(X):
    """
    http://docs.scipy.org/doc/numpy/reference/generated/numpy.linalg.svd.html
    Args:
        X ndarray(m, n)
    Returns:
        U ndarray(n, n): principle components
    """
    # 1. normalize data
    X_norm = normalize(X)

    # 2. calculate covariance matrix
    Sigma = covariance_matrix(X_norm)  # (n, n)

    # 3. do singular value decomposition
    # remeber, we feed cov matrix in SVD, since the cov matrix is symmetry
    # left sigular vector and right singular vector is the same, which means
    # U is V, so we could use either one to do dim reduction
    U, S, V = np.linalg.svd(Sigma)  # U: principle components (n, n)

    return U, S, V


def project_data(X, U, k):
    """
    Args:
        U (ndarray) (n, n)
    Return:
        projected X (n dim) at k dim
    """
    m, n = X.shape

    if k > n:
        raise ValueError('k should be lower dimension of n')

    return X @ U[:, :k]


def recover_data(Z, U):
    m, n = Z.shape

    if n >= U.shape[0]:
        raise ValueError('Z dimension is >= U, you should recover from lower dimension to higher')

    return Z @ U[:, :n].T

```

```python
mat = sio.loadmat('./data/ex7faces.mat')
mat.keys()
```

```python
X = np.array([x.reshape((32, 32)).T.reshape(1024) for x in mat.get('X')])
X.shape
```

```python
plot_n_image(X, n=64)
```

# run PCA, find principle components

```python
U, _, _ = pca(X)
```

```python
U.shape
```

```python
# didn't see face in principle components 
plot_n_image(U.T, n=36)
```

I don't see face here...

# reduce dimension to `k=100`

```python
# no face in projected dimension too
Z = project_data(X, U, k=100)
plot_n_image(Z, n=64)
```

# recover from `k=100`
you lost some detail, but.. they are strikingly similar

```python
X_recover = recover_data(Z, U)
plot_n_image(X_recover, n=64)
```

```python
def projectAndRecover(dimension):

    Z1 = project_data(X[0:1], U, k=dimension)
    X1_recover = recover_data(Z1, U)
    return X1_recover;

from time import *



def hist_similar(lh, rh):
    # assert len(lh) == len(rh)
    return sum(1 - (0 if l == r else float(abs(l - r))/max(l, r)) for l, r in zip(lh[0], rh[0]))/len(lh)


def calc_similar(li, ri):
    return hist_similar(li.histogram(), ri.histogram())




x=np.linspace(1,1023,100,dtype=int)
ytime=[]

xlist= x.tolist()
for i in xlist:
   # begin_time = time()
   recover= projectAndRecover(i)
   # end_time = time()
   # run_time = end_time-begin_time
   # ytime.append(run_time)

   similar=hist_similar(np.histogram(X[0],bins=23,range=(-130,100)),np.histogram(projectAndRecover(i)[0],bins=23,range=(-130,100)))
   ytime.append(similar)

plt.plot(x, ytime, label='y = sin(x)')
plt.show()
plt.axis('on')
```
