```python
%matplotlib inline
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt 
import scipy.io as sio
from PIL import Image
```

```python
def loadImage(path):
    img = Image.open(path)
    # 将图像转换成灰度图
    img = img.convert("L")
    # 图像的大小在size中是（宽，高）
    # 所以width取size的第一个值，height取第二个
    width = img.size[0]
    height = img.size[1]
    data = img.getdata()
    # 为了避免溢出，这里对数据进行一个缩放，缩小100倍
    data = np.array(data).reshape(height,width)/100
   

    return data
def pca(data,k):

    n_samples,n_features = data.shape
    # 求均值
    mean = np.array([np.mean(data[:,i]) for i in range(n_features)])
    # 去中心化
    normal_data = data - mean
    # 得到协方差矩阵
    matrix_ = np.dot(np.transpose(normal_data),normal_data)
    # 有时会出现复数特征值，导致无法继续计算，这里用了不同的图像，有时候会出现复数特征，但是经过sklearn中的pca处理同样图像可以得到结果
    
    eig_val,eig_vec = np.linalg.eig(matrix_)

    # 第二种求前k个向量
    eigIndex = np.argsort(eig_val)
    eigVecIndex = eigIndex[:-(k+1):-1]
    feature = eig_vec[:,eigVecIndex]
    new_data = np.dot(normal_data,feature)
    # 将降维后的数据映射回原空间
    rec_data = np.dot(new_data,np.transpose(feature))+ mean
    # print(rec_data)
    return rec_data
def show(newImage):
    plt.imshow(newImage)
    plt.axis('on') # 关掉坐标轴为 off
    plt.title('image') # 图像题目
    plt.show()
```

```python
imgdata=loadImage("dog.jpg")
# 查看原图的话，需要还原数据
new_im = Image.fromarray(imgdata*100)
show(new_im)
```

```python
rec_data=pca(imgdata,100)
# 压缩后的数据也需要乘100还原成RGB值的范围
newImage = Image.fromarray(rec_data*100)
show(newImage)
```

# 设置不同的k


```python
fig,ax_array=plt.subplots(nrows=2,ncols=2,figsize=(8,8),dpi=100);
kList=[10,50,100,400]
for i in range(2):
    for j in range(2):
        img=pca(imgdata,kList[i*2+j])
        ax_array[i,j].imshow(img,cmap="gray")
        ax_array[i,j].set_title("set k = %d " % kList[i*2+j])
        ax_array[i,j].axis('off')

##设定子图间距 ， left < right, top > bottom, 数字表示窗口大小的比例（如下则子图间距为窗口大小的1%）
plt.subplots_adjust(left=0.04, top= 0.96, right = 0.96, bottom = 0.04, wspace = 0.1, hspace = 0.1)
```

# 应用于人脸数据集


```python
mat = sio.loadmat('./data/ex7faces.mat')
mat.keys()
```

```python
X = np.array([x.reshape((32, 32)).T.reshape(1024) for x in mat.get('X')])
X.shape
```

```python
def plot_n_image(X, n):
    """ plot first n images
    n has to be a square number
    """
    pic_size = int(np.sqrt(X.shape[1]))
    grid_size = int(np.sqrt(n))

    first_n_images = X[:n, :]

    fig, ax_array = plt.subplots(nrows=grid_size, ncols=grid_size,
                                    sharey=True, sharex=True,figsize=(8, 8) )

    for r in range(grid_size):
        for c in range(grid_size):
            ax_array[r, c].imshow(first_n_images[grid_size * r + c].reshape((pic_size, pic_size)),cmap='gray')

            plt.xticks(np.array([]))
            plt.yticks(np.array([]))
```

```python
X64=X[:64,:]
```

```python
Y=np.zeros((64,32*32))
for i in range(X64.shape[0]):
    data=X64[i].reshape(32,32)
    Y[i]=pca(data,10).reshape(-1)
plot_n_image(Y, 64)
```

# 以第一张图片为例比较相似度


```python

def hist_similar(lh, rh):
    # assert len(lh) == len(rh)
    return sum(1 - (0 if l == r else float(abs(l - r))/max(l, r)) for l, r in zip(lh[0], rh[0]))/len(lh)
```

```python
x=np.linspace(1,31,31,dtype=int)
ytime=[]
xlist= x.tolist()
first=X64[0].reshape(32,32)
for i in xlist:
   similar=hist_similar(np.histogram(X[0],bins=23,range=(-130,100)),np.histogram(pca(first,i).reshape(-1),bins=23,range=(-130,100)))
   ytime.append(similar)
plt.plot(x, ytime, label='y = sin(x)')
plt.show()
plt.axis('on')

```
